```python
import tkinter as tk
from tkinter import ttk
from tqdm import tqdm
import time
import glob
import pandas as pd

# Suppress pandas warnings
pd.options.mode.chained_assignment = None
```

```python
files = glob.glob('*.xlsx')
```

```python
sheet_name_list = [
                    '結果_1', '結果_2', '結果_3', '結果_4', '結果_5', '結果_6',
                    '結果_7', '結果_8', '結果_9', '結果_10', '結果_11', '結果_12',
                    ]
```

```python
def data_concat(files):
    df_list = []
    result_list = []
    for file in tqdm(files):
        for sheet_name in sheet_name_list:
            id = file.split('.')[0]+'_'+sheet_name
            df = pd.read_excel(file, header=1, sheet_name=sheet_name)
            if '総合 値' in df.columns:
                df = df.rename(columns={'総合 値':'総合値'})
            if 'Unnamed: 10' in df.columns:
                df = df.drop(
                    ['Unnamed: 10'],axis=1
                    )
            df_drop = df.dropna(subset=['着順'])
            try:
                df_drop = df_drop.drop(
                                    [
                                    '着順','馬名','騎手名', '前回騎乗', '調教師'
                                    ],
                                    axis=1
                                    )
            except KeyError:
                df_drop = df_drop.drop(
                                    [
                                    '着順','馬名','騎手名', '調教師'
                                    ],
                                    axis=1
                                    ) 
            result = df.query(str(df.columns[5])+' == "3連複"')[[
                                                    str(df.columns[5]), str(df.columns[6]),str(df.columns[7])
                                                    ]]
            result = result.rename(columns={
                                    str(df.columns[5]):'券種',
                                    str(df.columns[6]):'買い目',
                                    str(df.columns[7]):'odds'
                                    },)
            df_drop['id'] = id
            df_drop['race'] = id.split('_')[-1]
            result['id'] = id
            df_column = df_drop
            result_list.append(result)
            df_list.append(df_column)
    return pd.concat(df_list).to_csv('data_concat.csv',index=False, encoding='cp932'),\
            pd.concat(result_list).to_csv('result_concat.csv',index=False, encoding='cp932')
```

```python


class TkinterTqdm(tqdm):
    def __init__(self, *args, **kwargs):
        self.root = kwargs.pop("root")
        self.progressbar = kwargs.pop("progressbar")
        self.time_label = kwargs.pop("time_label")
        super().__init__(*args, **kwargs)
        self.start_time = time.time()

    def update(self, n=1):
        super().update(n)
        self.progressbar['value'] = self.n / self.total * 100
        
        elapsed_time = time.time() - self.start_time
        remaining_time = (self.total - self.n) / self.n * elapsed_time if self.n > 0 else 0
        
        self.time_label.config(text=f"経過時間: {elapsed_time:.2f}s 残り時間: {remaining_time:.2f}s")
        self.root.update_idletasks()

def main():
    root = tk.Tk()
    root.title("tkinter tqdm Example")

    progressbar = ttk.Progressbar(root, orient="horizontal", length=300, mode="determinate")
    progressbar.pack(pady=20)

    time_label = tk.Label(root, text="経過時間: 0.00s 残り時間: 0.00s")
    time_label.pack(pady=10)

    start_button = tk.Button(root, text="Start", command=lambda: data_concat(files))
    start_button.pack(pady=10)

    root.mainloop()

if __name__ == "__main__":
    main()
```
